```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('Solarize_Light2')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Red')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,8))
```
