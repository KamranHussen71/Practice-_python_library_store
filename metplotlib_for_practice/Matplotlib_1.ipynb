# Bar plot

```python
import matplotlib.pyplot as plt

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("Language",fontsize=10)
plt.ylabel("Number",fontsize=10)
plt.title("Graph",fontsize=10)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color="r",label="popularity")
plt.bar(x,z,width=0.4,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Language",fontsize=10)
plt.ylabel("Number",fontsize=10)
plt.title("Graph",fontsize=10)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")

plt.xticks(p+width/2,x,rotation=20)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("Language",fontsize=10)
plt.ylabel("Number",fontsize=10)
plt.title("Graph",fontsize=10)

plt.barh(x,y,color="r",label="popularity")
plt.barh(x,z,color="y",label="popularity1")

plt.legend()
plt.show()
```
