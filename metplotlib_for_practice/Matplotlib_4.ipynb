# Pie Plot

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex = [0.3,0.0,0.0,0.0]
c = ["r","b","g","y"]
plt.pie(x, labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,startangle=0
        ,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5},)

plt.title("Learning")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```
