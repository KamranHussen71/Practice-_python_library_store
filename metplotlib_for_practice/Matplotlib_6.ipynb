# Box Plot

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x, tick_labels=["Python"],showmeans=True, sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="r"),
           flierprops=dict(markeredgecolor="y"))

plt.show()
```

```python
x = [10,20,30,40,50,60,70]
x1 = [10,20,30,40,50,60,70,80,90]

y = [x,x1]
plt.boxplot(y, tick_labels=["Python","C++"],showmeans=True, sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="r"),
           flierprops=dict(markeredgecolor="y"))

plt.show()
```
