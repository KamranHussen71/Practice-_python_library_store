# HeaatMap Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap="PuOr",annot=True,)
```

```python
data = sns.load_dataset("anagrams")
x = data.drop(columns=["attnr"],axis=1).head(10)
x 
```

```python
sns.heatmap(x,annot=True)
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
y = {"fontsize":20,"color":"r"}
v = sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidth=1,linecolor="y",
            cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="Python",ylabel="learning")
sns.set(font_scale=5)
plt.show()
```

```python
ar = np.array([["Kamran","Nawshad","Rafid","Sanim","Tushar"],["Rahim","Karim","Jamil","Monir","Sakir"]])
ar
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=ar,fmt="s")
```
