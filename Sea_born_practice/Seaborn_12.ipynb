# Styling plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set_style("white")

sns.set_context("paper",font_scale=1)

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
sns.set_style("white")

sns.barplot(x="day",y="total_bill",data=var)

sns.despine()
plt.show()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(12,5))

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```
