# Strip Plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill", data=var,hue="sex",
              palette="Accent_r",linewidth=1,edgecolor="g"
              ,size=5,alpha=0.5)
plt.legend(loc=2)
plt.show()
```

```python
sns.stripplot(y=var["total_bill"])
plt.show()
```
