# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,order=["Fri","Sun","Thur","Sat",],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"g"},
           linewidth=3,)

plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="total_bill",y="day",data=var,order=["Fri","Sun","Thur","Sat",],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"g"},
           linewidth=3,)

plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(data=var,
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"g"},
           linewidth=3,orient="h")

plt.show()
```
