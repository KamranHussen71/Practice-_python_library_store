# Violine Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var,linewidth=2,color="g")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var,order=["Dinner","Lunch"],inner="box")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=var, hue="sex",)
plt.show()
```

```python
sns.violinplot(y=var["total_bill"])
plt.show()
```
