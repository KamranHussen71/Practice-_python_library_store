# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins").head(100)
var
```

```python
m = {"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(100,100),
                palette="gist_ncar",alpha=1,markers=m)
plt.legend(loc=4)
plt.show()
```
